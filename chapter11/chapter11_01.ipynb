# Chapter 11. 나이브 베이지안 분류기
## 1. 베이즈 정리의 이해
### 01 - 확률의 표현
- 이산형 값
    - 전체 표본공간에서 가능한 모든 결과 횟수 대비 해당 사건의 결과 횟수에 대한 비율을 나타냄
        - $P(X) = \frac{count(Event_x)}{count(Event_{allevent})}$
    - 이러한 이산형 값에 대한 분포를 표현하기 위해 먼저 값의 구간을 나눈 다음 해당 구간에 포함되는 값의 개수를 나타내는 도수분포표 사용
- 연속형
    - 해당 데이터를 적절히 표현하는 함수를 생성한 후 해당 함수의 적분값을 취함
    - 이산형 값과는 달리 특정 위치의 값은 없고 구간의 값(해당 함수의 적분값)을 취해 구간의 확률 계산
    - (연속형 값의 경우 확률로 계산하는 것이 매우 난해 => 교재에선 먼저 이산형 값의 확률을 다룸)

#### 1.1 확률의 기본 성질
- 확률은 다음 수식과 같이 모든 사건에 대해 반드시 0 ~ 1 사이의 값을 가져야 함
    - $0 \le P(X) \le 1$
- 또한 각 사건들이 서로 관계가 없는 경우(각 사건들이 일어날 확률이 다른 사건이 일어날 확률에 영향 X)
    - 각 사건들이 **독립**되었다고 함
    - 각 사건이 독립일 때 모든 사건의 확률 합은 1
        - $P(S) = \displaystyle\sum_{i=1}^N P(E_i) = 1$

#### 1.2 조건부 확률
- 어떤 사건이 일어난다 가정했을 때 다른 사건이 일어날 확률
    - $P(A | B) = \frac{P(A\cap B)}{P(B)}$

### 02 - 베이즈 정리
- **베이즈 정리(Bayes' theorem)** : 두 확률 변수의 사전확률과 사후확률 사이의 관계를 나타내는 정리
    - 기본 전제 : 객관적인 확률이 존재하지 않는다는 것
        - 확률은 이전에 일어난 사건의 연속이므로 이 사건으로 인해 확률이 지속적으로 업데이트 된다는 뜻

- 베이즈 정리는 빈도주의자(frequentist)와 베이즈주의자(Bayesian) 간의 논쟁으로 많이 설명됨
    - 두 이론을 비교해보면 확률 자체의 계산법은 동일하지만 해석하는 방식의 차이가 다름
        - 빈도주의자 : 세상의 모든문제에 이미 일어난 많은 사건을 바탕으로 해당 확률이 발생
        - 베이즈주의자 : 확률이라는 것은 사람들의 믿음에서 발생
        - 예시
            - 3장의 카드 중 원하는 카드를 찾을 확률은 상식적으로 $\frac{1}{3}$
            - 실제로 카드를 뒤집었더니 각 카드마다 조금씩 다른 확률이 나오게 됨
                - 베이즈주의자는 이 확률을 기반으로 실행할 때마다 계속하여 확률들을 업데이트
                - 빈도주의자는 다음 확률을 계속하여 실행할 경우 $\frac{1}{3}$에 수렴할 것이므로 $\frac{1}{3}$로 간주

- 베이즈주의를 이해하기 위해 가장 기본적으로 알아야 하는 개념은 사전확률과 사후확률
    - 이를 위해 가장 기본적인 베이즈 정리의 수식 사용
        - $P(H | D) = \frac{P(H)P(D | H)}{P(D)}$
        - 이 수식에서 $H$는 가설, $D$는 데이터를 의미
            - $P(H | D)$는 사후확률
    - 사후확률 : 데이터가 주어졌을 때 해당 가설이 맞는지에 대한 확률
        - 사전에 발생한 사전확률($P(H)$)에 가능도(likelihood : $P(D | H)$)를 곱해서 데이터의 확률($P(D)$)로 나눠 구함
    - 가능도(likelihood) : 어떤 사건이 발생했을 때 다음 사건이 발생할 수 있는 모든 확률발생가능한 정도를 확률로 나타낸 수치
        - $P(D | H)$의 의미 : 가설이 주어졌을 때 해당 데이터가 존재할 확률
        - 가능도는 일반적으로 분포의 값으로 표현할 수 있음 (H 가설의 종류에 따라 여러 확률로 나타낼 수 있음)
    - $P(D)$ : 해당 데이터가 발생한 확률
        - 일반적으로 나이브 베이지안 분류에서 $P(D)$는 모든 가설 $H$에 대해 동일함
            - 그래서 고려하지 않을 때가 많음
