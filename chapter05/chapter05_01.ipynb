```python
# 데이터 시각화(data visualization) : 데이터 분석 결과를 시각적으로 표현하고 전달하는 과정

# 파이썬의 시각화 도구는 matplotlib, seaborn, plotly(, bokeh, pygal, ggplot)등이 있음
    # 책에선 matplotlib, seaborn이 중심 (pandas와 연동 가능)
        # matplotlib : 파이썬의 표준 시각화 도구, 다른 도구들의 가장 기본이 되는 시각화 분석 도구
        # seaborn : matplotlib을 기반으로 한 시각화 도구, matplotlib을 쉽게 사용하기 위한 함수 지원
```

```python
# matplotlib의 구조
    # 기본적으로 matlab에서 사용하는 기능들을 파이썬에서 사용할 수 있게 해주는 시각화 도구
        # 엑셀에서 사용하는 차트, 그래프를 작성할 수 있게 해줌
        # 다양한 함수 지원 -> 사용자가 맞춤형 그래프를 작성할 수 있도록 함

import matplotlib.pyplot as plt
# matplotlib 모듈 실행 문구
```

```python
# pyplot (파이플롯)
    # matplotlib을 이용할 때 가장 기본이 되는 객체, 구조
    # 그래프를 작성할 수 있게 준비해주는 화방과 같은 역할 수행
    # 다양한 그림을 그리는 도구들을 사용할 수 있게 함
    # 그림(figure)은 pyplot 위에 올려진 그림판, 축(axes)은 최종적으로 그림판 위에 올려진 도화지 같은 존재
        # 여러 장의 axes를 figure 위에 올린 다음 여러 그래프를 작성할 수 있음
```

```python
# 가장 간단한 그래프 작성 코드
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X, Y)
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
# 두 그래프가 하나의 figure 위에 그려짐
```

```python
# figure과 axes
    # subplots 함수 사용
        # plot 함수를 사용하면 한 번에 그래프를 작성하지만, subplots는 그림 객체와 축 객체를 나눠 받음

fig, ax = plt.subplots()                # fig, ax에 그림과 축 객체를 할당
x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]

ax.plot(x_1, y_1)                       # 축 객체에 plot 함수를 사용해서 그래프 생성
ax.set(title = "cos graph",             # 축 위에 그래프의 제목, x축과 y축의 라벨 설정
       xlabel = 'X',
       ylabel = 'Y')

plt.show()
```

```python
# subplot 함수 이해하기
fig = plt.figure()              # figure 반환
fig.set_size_inches(10, 10)     # figure의 크기 지정

ax_1 = fig.add_subplot(1, 2, 1) # 첫 번째 그래프 생성 (행은 1개, 열은 2개로 그림 객체 공간을 지정)
ax_2 = fig.add_subplot(1, 2, 2) # 두 번째 그래프 생성 (첫 번째 축 겍체를 ax_1, 두 번째 열에는 ax_2라는 이름으로 축 객체 생성)

ax_1.plot(X_1, Y_1, c = 'b')    # 첫 번째 그래프 설정
ax_2.plot(X_2, Y_2, c = 'g')    # 두 번째 그래프 설정
plt.show()                      # 그래프 출력
```

```python
# 서브플롯 행렬
    # 축을 여러 개 만들 때 일반적으로 그림객체에서 add_subplot 함수를 사용 or plot 객체에서 subplots 함수를 사용하여 먼저 축 객체 공간을 확보해야 함

fig, ax = plt.subplots(nrows= 1, ncols= 2)  # 행은 1개, 열은 2개인 그림(figure) 객체 공간을 생성하고 이를 하나의 행렬(matrix)처럼 생각하여 두 개의 축(axes) 객체가 생성됨
print(ax)
# 다음 코드를 실행하면 ax 변수에는 axes 객체가 넘파이 배열 타입으로 생성됨
```

```python
# 아래 코드는 subplots 함수에서 2x2 행렬을 가진 그림 객체가 생성되고 여기에 넘파이 배열의 인덱스를 사용하여 각 축 객체에 접근한다는 뜻
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, y_1)
ax[0, 1].plot(x, y_2)
ax[1, 0].plot(x, y_3)
ax[1, 1].plot(x, y_4)

plt.show()
```

```python
# matplotlib은 행과 열을 지정한 후, 그 다음 번호는 행과 열이 지정된 행렬에서 차례대로 축의 위치를 배열한다
    # 2x3 크기의 그림객체가 있을 때, 첫 행에는 1, 2, 3의 그래프, 두 번째 행에서는 4, 5, 6의 그래프가 들어감
        # (2, 3, 1), (2, 3, 2), (2, 3, 3)
        # (2, 3, 4), (2, 3, 5), (2, 3, 6)

# 행렬 형태로 그래프를 그리는 방식 외에도 다소 복잡한 형태의 그래프 생성도 가능

ax1 = plt.subplot(321)  # 첫 번째 공간에 axes 생성
plt.plot(x, y_1)
ax2 = plt.subplot(322)  # 두 번째 공간에 axes 생성
plt.plot(x, y_1)
ax3 = plt.subplot(312)  # 두 번째 공간에 axes 생성
plt.plot(x, y_1)
ax4 = plt.subplot(325)  # 다섯 번째 공간에 axes 생성
plt.plot(x, y_1)
ax5 = plt.subplot(326)  # 여섯 번째 공간에 axes 생성
plt.plot(x, y_1)
# 321의 경우 3x2의 공간을 만든 후, 첫 번째 칸에 그림을 넣는 것
# 312의 경우 3x1의 공간을 만든 후, 두 번째 칸에 그림을 넣는 것

plt.show()
```

```python
# matplotlib으로 그래프 꾸미기
    # 색상
        # color 또는 c 매개변수를 새ㅏ용해서 색상 변경 가능
            # RGB 값을 사용해서 '#'을 붙이고 16진법으로 색상 표현
            # 규칙이 있는 약어 입력(b, g, r, c, m, y, k, w 등)

X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color= "#000000")
plt.plot(X_2, Y_2, c= 'c')

plt.show()
```

```python
    # 선의 형태
        # linestyle 또는 ls 매개변수 사용
            # 선의 형태는 점선 형태인 dashed나 직선 형태인 solid 등이 있음

plt.plot(X_1, Y_1, c= 'b', linestyle= "dashed")
plt.plot(X_2, Y_2, color= 'r', ls= "dotted")

plt.show()
```

```python
    # 제목
        # 축 객체마다 제목을 달 수 있음

plt.plot(X_1, Y_1, c= 'b', linestyle= "dashed")
plt.plot(X_2, Y_2, color= 'r', ls= "dotted")

plt.title("Two Lines")
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c= 'b')
ax_1.set_title("Figure 1")
ax_2.plot(X_2, Y_2, c= 'g')
ax_2.set_title("Figure 2")

plt.show()
```

```python
    # 범례
        # 그래프에 대한 설명인 범례도 축별로 설정 가능
            # legend 함수 사용하면 범례가 생김
            # 범례를 꾸미기 위해 사용되는 매개변수는 shadow, 범례 뒤 쪽에 그림자 효과 추가 가능
            # loc는 범례가 위치하는 곳을 지정하는 매개변수(center, upper right 등 총 11개 값 입력 가능)
                # best를 지정하면 적절한 위치에 범례를 위치시킴

plt.plot(X_1, Y_1, color= 'b', linestyle= "dashed", label= "line_1")
plt.plot(X_2, Y_2,
         color= 'r',
         linestyle= "dotted",
         label= "line 2")

plt.legend(
    shadow= "True",
    fancybox= "False",
    loc= "upper right")

plt.title("$y = ax + b$")   # $ 에 따라 표기 방식이 달라짐
plt.xlabel("$x_line$")
plt.ylabel("y_line")

plt.show()
```

```python
# matplotlib에서 사용하는 그래프
    # matplotlib은 기본적으로 그래프를 꾸며주는 공통 도구를 제외하고 각 데이터의 특성에 맞는 다양한 그래프 지원

    # 산점도 그래프
        # 산점도(scatter plot)는 일반적으로 데이터의 분포를 2차원 평면에 표현할 때 사용하는 그래프
            # 가장 간단한 형태는 x, y에 각각 좌표에 대한 정보를 담고 이를 scatter 함수를 사용하여 figure에 생성하는 것
        # 매개변수 c는 색상, marker는 각 포인트의 모양을 지정

data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:, 0],
            data_1[:, 1],
            c= 'b', marker= 'x')
plt.scatter(data_2[:, 0],
            data_2[:, 1],
            c= 'r', marker= 'o')

plt.show()
```

```python
# 추가적으로 size 매개변수를 이용하면 각 포인트(point)의 크기도 지정 가능

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (
    15 * np.random.rand(N) ** 2)

plt.scatter(x, y,
            s= area, c= colors, # s는 size 매개변수
            alpha= 0.5)

plt.show()
```

```python
    # 막대 그래프
        # 막대 그래프(bar graph)는 일반적으로 데이터의 개수나 크기를 비교할 때 사용하는 그래프

data =[[5., 25., 50., 20.],
       [4., 23., 51., 17],
       [6., 22., 52., 19]]
# 데이터 생성

X = np.arange(0, 8, 2)
# X 좌표 시작점

plt.bar(X + 0.00, data[0], color= 'b', width= 0.50)
plt.bar(X + 0.50, data[1], color= 'g', width= 0.50)
plt.bar(X + 1.0, data[2], color= 'r', width= 0.50)
# 3개의 막대 그래프 생성

plt.xticks(X + 0.50, ('A', 'B', 'C', 'D'))
# X축에 표시될 이름과 위치 설정

plt.show()
```

```python
    # 누적 막대 그래프
        # 누적 막대 그래프(stacked bar graph)는 데이터를 밑에서부터 쌓아올려 표현할 때 사용
        # 아래 코드에서 변수 data에 있는 값은 옆으로 데이터를 추가하는 것이 아닌 위로 데이터를 추가하는 형태
        # for문을 사용하면 루프가 작동하면서 i의 값, 여기에서는 막대 상에서 아랫부분까지의 값을 더한 다음 그 값을 bottom에 넣으면 됨
data =np.array([[5., 25., 50., 20.],
                [4., 23., 51., 17],
                [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ['A', 'B', 'C']
X = np.arange(data.shape[1])

for i in range(3) :
     plt.bar(X, data[i],
            bottom= np.sum(data[:i], axis= 0),
            color = color_list[i],
            label = data_label[i])

plt.legend()
plt.show()
```

```python
    # 히스토그램
        # 히스토그램(histogram)은 데이터의 분포를 표현하는 그래프

N = 1000
X = np.random.normal(size= N)
# 랜덤한 데이터 생성

plt.hist(X, bins= 100)
# 매개변수 bins로 히스토그램에서 생기는 막대의 개수를 지정하면 그 개수만큼의 그래프가 작성됨
plt.show()
```

```python
    # 상자그림
        # 상자그림(boxplot)은 데이터의 분포를 나타내면서 밀집 정도를 표현하는 그래프
        # 상자그림은 사분위수를 시각화하여 그래프 형태로 나타냄
        # 1사분위수(Q1, 25%), 3사분위수(Q3, 75%)의 차이가 IQR(InterQuartile Range), 이 IQR과 Q1, Q3를 이용하여 최솟값, 최댓값과 이상치를 정함

data = np.random.randn(100, 5)
plt.boxplot(data)
plt.show()
```
