```python
# 시본의 기본
# 시본(seaborn)은 matplotlib을 바탕으로 다양한 함수를 사용하도록 도와주는 일종의 래퍼(wrapper) 모듈
    # 결과물은 맷플롯립과 동일하게 나옴
        # 그래프를 만드는 과정과 코드 작성이 매우 간단, 직감적이라는 장점
    # import seaborn as sns
```

```python
# 판다스의 데이터프레임을 활용 -> 비교적 편하게 그래프 작성 가능

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fmri = sns.load_dataset("fmri")                         # fmri 데이터셋 사용
sns.set_style("whitegrid")                              # 기본 스타일 적용
sns.lineplot(x= "timepoint", y= "signal", data= fmri)   # 선그래프 작성

# 시본에서 자체적으로 제공하는 데이터셋을 사용. 여기서는 fmri 데이터셋을 사용
# whitegrid 스타일은 하얀색의 그리드가 생성되는 스타일
# 선그래프(lineplot) 작성. 매개변수 x와 y에는 각 열의 이름 입력, 매개변수 data에는 데이터프레임 객체 이름 지정
```

```python
# 시본은 맷플롯립의 그림, 축 객체와 같은 복잡한 개념 없이 단순하게 그래프가 생성됨
    # xticks의 설정 없이도 각 축에 라벨이 자동으로 생성됨
        # 데이터프레임과 x, y에 해당하는 열 이름만 지정하면 비교적 정확히 그래프를 그릴 수 있음

# 실제 fmri 데이터는 연속형 값 뿐만 아니라 다양한 범주형 값도 가지고 있음
    # 맷플롯립을 사용하면 복잡하게 표현해야 함
        # 시본은 단순이 hue라는 매개변수만 추가함으로써 꽤 그럴듯한 그래프 작성 가능

fmri.sample(n= 10, random_state= 1)
```

```python
sns.lineplot(x= "timepoint", y= "signal", hue= "event", data= fmri)

# 이 코드의 위에 있는 코드를 보면 데이터를 먼저 확인함
    # 이 데이터에서 event는 범주형 변수로 다른 데이터들과 구분할 때 사용이 가능함
        # 즉 timepoint에 따른 signal 값을 선그래프로 보여주면서 event를 기준으로 구분하려면 이 코드와 같이 사용 가능
```

```python
# 시본에서 사용하는 그래프
    # 1. 회귀 그래프
        # 회귀 그래프(regression plot)는 회귀식을 적용하여 선형회귀 추세선을 그래프에 함께 작성하는 것
        # 함수로 regplot 사용
    # 다음 코드에 나오는 x_ci는 신뢰구간의 비율을 나타내는 매개변수
        # 선형회귀 추세선은 데이터를 기반으로 데이터의 x값 대비 y값의 변화를 예측하는 직선

tips = sns.load_dataset("tips")
sns.regplot(x= "total_bill", y= "tip", data= tips, x_ci= 95)
```

```python
    # 2. 산점도
        # 산점도(scatter plot)는 x와 y를 기준으로 데이터의 분포를 표현하는 그래프
        # 함수로 scatterplot 사용
    # 다음 코드에서 x와 y열에 들어갈 수 있는 열 이름들을 넣으면 자동으로 산점도를 작성할 수 있음

tips = sns.load_dataset("tips")
sns.scatterplot(x= "total_bill", y= "tip", hue= "time", data= tips)
```

```python
    # 3. 비교 그래프
        # 비교 그래프(counter plot)는 일반적으로 범주형 데이터의 항목별 개수를 그래프로 표현하는 것

tips = sns.load_dataset("tips")
sns.countplot(x= "smoker", hue= "time", data= tips)
```

```python
    # 4. 막대그래프
        # 시본의 막대그래프(bar plot)는 barplot 함수 사용 -> 다소 특이함
        # lineplot 함수를 사용해도 데이터의 평균값의 변화를 나타낼 수 있음
            # 일반적인 막대그래프는 데이터의 개수를 표시하지만, 시본은 y값이 연속형 값일 경우 평균값을 나타냄
    # 다음 코드의 실행결과에서 나타난 그래프 형태와 같이 해당 데이터의 신뢰구간을 검은색 막대로 표시

sns.barplot(x= "day", y= "total_bill", data= tips)
```

```python
# 사전 정의된 그래프
    # 사전 정의된 그래프 : 맷플롯립 관점에서 하나 이상의 그래프들을 합쳐 다양한 정보를 줄 수 있는 그래프
        # 범주형 데이터에서 매우 유용하게 쓸 수 있는 그래프들이 많음
```

```python
    # 1. 분포를 나타내는 그래프 : 바이올린 플롯과 스웜 플롯
        # 바이올린 플롯(violin plot) : 데이터의 분포를 바이올린과 비슷한 형태로 보여주는 그래프
            # 상자그림과 분포도를 한 번에 나타낼 수 있음
            # violinplot 함수를 사용하여 x축에는 범주형, y축에는 연속형 데이터를 넣어 사용

sns.violinplot(x= "day", y= "total_bill", hue= "smoker", data= tips, palette= "muted")
```

```python
        # 스웜 플롯(swarm plot) : 바이올린 플롯과 같은 형태지만, 산점도를 이용해서 데이터의 분포를 나타냄
            # 매개변수 hue와 함께 사용 -> 두 범주형 데이터를 비교할 때, 데이터의 형태를 점이 겹치지 않게 정리
                # 각 데이터의 영역별로 얼마나 많은 데이터가 있는지 직관적으로 파악하도록 도와줌

sns.swarmplot(x= "day", y= "total_bill", hue= "smoker", data= tips, palette= "muted")
                
```

```python
    # 2. 다양한 범주형 데이터를 나타내는 패싯그리드
        # 시본에서는 맷플롯립과 달리 그래프의 틀만 제공하여 적당한 그래프를 그려줄 수 있음
            # 이를 위해 패싯그리드라고 부르는 FacetGrid 클래스를 제공

sns.FacetGrid(tips, col= "time", row= "sex")
# 기본적으로 매개변수 col과 row에 범주형 데이터를 넣으면 해당 열의 데이터 종류만큼 m*n의 그래프 틀이 생성됨
```

```python
        # 이러한 그리드가 생성된 후 실제 그래프를 만들 때는 일종의 맵(map) 개념을 사용
            # 생성된 그래프를 각 FacetGrid에 있는 개별 그래프 영역에 집어넣는 구조로 되어있음

g = sns.FacetGrid(tips, col= "time", row= "sex")

g.map(sns.scatterplot, "total_bill", "tip")
# 이 코드를 보면 time과 sex내에 있는 범주형 데이터의 종류에 따라 데이터를 나눔
    # 그 다음 산점도 함수 scatterplot의 x축, y축에 total_bill과 tip값을 대응시킴
```

```python
        # 만약 다른 그래프로 작성하고 싶다면, map_dataframe을 사용
            # histplot함수에 적절한 매개변수를 지정하여 데이터를 넘겨줌

g = sns.FacetGrid(tips, col= "time", row= "sex")

g.map_dataframe(sns.histplot, x= "total_bill")
```
